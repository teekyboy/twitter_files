```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("small_sample.csv")
df.columns
```

```python
df['prof_1min'] = np.random.uniform(-20, 20, size=len(df))
df['prof_60min'] = np.random.uniform(-20, 20, size=len(df))
df['prof_1day'] = np.random.uniform(-20, 20, size=len(df))
df['prof_1week'] = np.random.uniform(-20, 20, size=len(df))
```

```python
df
```

```python
df = df.dropna(subset=['cashtags'])
```

```python
np.unique(df["cashtags"].values)
```

```python
import re
def clean_cashtgs(st):
    letters = re.findall('[A-Za-z]+', st)
    return ' '.join(letters)
```

```python
clean_cashtgs("['MGNI', 'USER', 'TGTX', 'GLYC', 'GRWG', 'APPS', 'BAND', 'VAPO', 'UFAB', 'CLVS', 'RPAY', 'EVER', 'RDFN', 'TLS', 'OPRT', 'RYAM', 'NEO', 'APYX', 'TSP']")
```

```python
df["cashtags"] = df["cashtags"].apply(clean_cashtgs)
```
